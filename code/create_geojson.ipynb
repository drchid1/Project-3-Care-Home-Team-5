```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/CareHome-data.csv', delimiter=',')
```

```python
df.head()
```

```python
# Function to make the header row all into camelCase
def to_camel_case(s):
    s = s.title().replace(' ', '')
    return s[0].lower() + s[1:]

# Apply the function to the headers and remove '\n' and '&'
df.columns = df.columns.str.replace('&', '')
df.columns = df.columns.str.replace('\n', '')
df.columns = df.columns.map(to_camel_case)
```

```python
json_result_string = df.to_json(
    orient='records', 
    double_precision=12,
    date_format='iso'
)
json_result = json.loads(json_result_string)
```

```python
geojson = {
    'type': 'FeatureCollection',
    'features': []
}
for record in json_result:
    geojson['features'].append({
        'type': 'Feature',
        'geometry': {
            'type': 'Point',
            'coordinates': [record['latitude'], record['longitude']],
        },
        'properties': record,
    })
```

```python
with open('../data/ch_coordinates.json', 'w') as f:
    f.write(json.dumps(geojson, indent=2))
```
