```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/nhomes_geocoded.csv', delimiter=',')
```

```python
df.head()
```

```python
json_result_string = df.to_json(
    orient='records', 
    double_precision=12,
    date_format='iso'
)
json_result = json.loads(json_result_string)
```

```python
geojson = {
    'type': 'FeatureCollection',
    'features': []
}
for record in json_result:
    geojson['features'].append({
        'type': 'Feature',
        'geometry': {
            'type': 'Point',
            'coordinates': [record['latitude'], record['longitude']],
        },
        'properties': record,
    })
```

```python
with open('../data/geojson_coordinates.json', 'w') as f:
    f.write(json.dumps(geojson, indent=2))
```
