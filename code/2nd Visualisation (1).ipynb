```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
file_path = "C:/Users/reach/Downloads/final_nh_data_with_ratings (1).xlsx"
df = pd.read_excel(file_path)

# Check the first few rows to understand the structure of the data
df.head()

```

```python
# Ensure the correct data types for each column
df['Name & Address'] = df['Name & Address'].astype(str)
df['Email'] = df['Email'].astype(str)
df['Telephone No'] = df['Telephone No'].astype(str)
df['Age Range'] = df['Age Range'].astype(str)
df['CQC Registered For'] = df['CQC Registered For'].astype(str)



df['Capacity'] = pd.to_numeric(df['Capacity'], errors='coerce')
df['Post Code'] = df['Post Code'].astype(str)
df['Care Home Name'] = df['Care Home Name'].astype(str)
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
df['Rating'] = df['Rating'].astype(str)

# Handle missing data by filling or dropping
df.fillna({
    'Capacity': 0,
    'Latitude': 0,
    'Longitude': 0
}, inplace=True)

```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Capacity'], bins=30)
plt.title('Distribution of Care Home Capacities')
plt.xlabel('Capacity')
plt.ylabel('Count')

plt.show()
```

```python
# Calculate the counts of each age range
age_range_counts = df['Age Range'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(age_range_counts.index, age_range_counts.values)
plt.title('Distribution of Age Ranges in Care Homes')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Ratings Distribution with different colors for each bar
rating_counts = df['Rating'].value_counts()
colors = ['skyblue', 'salmon', 'lightgreen', 'gold', 'orange']  # specify colors for each rating

plt.figure(figsize=(10, 6))
plt.bar(rating_counts.index, rating_counts.values, color=colors[:len(rating_counts)])
plt.title('Distribution of Care Home Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. CQC Registered Services
services_counts = df['CQC Registered For'].value_counts().head(10)  # Select the first 10 categories
plt.figure(figsize=(10, 6))
plt.bar(services_counts.index, services_counts.values)
plt.title('CQC Registered Services')
plt.xlabel('Service')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels to be vertical
plt.show()
```
