```python
import pandas as pd
import numpy as np

#set the max number of rows to display
pd.set_option('display.max_rows', 100)
```

```python
# Open the dataframe from the csv file
df = pd.read_csv('../data/nhomes_geocoded.csv')
```

```python
df.head(100)
```

```python
capacity_df = pd.DataFrame()
```

```python
df.info()
```

```python
# Create a new dataframe with the care home name and the capacity
capacity_df = pd.DataFrame()

# Creating the new columns
capacity_df['care_home_name'] = df['careHomeName']
capacity_df['capacity'] = df['capacity']
capacity_df['capacity'] = df['capacity'].fillna(0)
capacity_df['capacity'] = capacity_df['capacity'].astype(int)
capacity_df['available_beds'] = 0

# Create random bed avilablity for each care home
for rec in range(0, len(capacity_df)):
    if capacity_df.loc[rec, 'capacity'] != 0:
        available_beds = np.random.randint(0, capacity_df.loc[rec, 'capacity'])
        available_beds *= 0.2
        available_beds = int(available_beds)
    else:
        available_beds = 0
    capacity_df.loc[rec, 'available_beds'] = available_beds
```

```python
# Save the datafame to an excel file
capacity_df.to_excel('../data/nhomes_capacity.xlsx', index=False)
```
